```python
import pandas as pd
```

```python
dados = pd.read_csv('Dados/aluguel.csv', sep = ';')
dados.head()
```

```python
 # Contar a quantidade de imóveis que possuem 1 e 2 quartos, 3 e 4 , 5 e 6 , 7 ou mais quartos
classe = [0, 2, 4, 6, 100]
```

```python
quartos = pd.cut(dados.Quartos, classe)
quartos
# A função cut() é uma ferramenta do pandas que auxilia na criação distribuições de frequências.
```

```python
pd.value_counts(quartos)
# ( -> intervalo aberto valor não entra
# ] -> intervalo fechado valor entra
```

```python
labels = ['1 e 2 quartos', '3 e 4 quartos', '5 e 6 quartos', '7 quartos ou mais']
quartos = pd.cut(dados.Quartos, classe, labels = labels)
pd.value_counts(quartos)
```
